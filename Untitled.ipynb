```python
with open("words.txt", "r") as file:
    data = file.read()
words = data.split('\n')
```

```python
new_words = list()
for word in words:
    new_words.append(word.upper())
new_words

```

```python
import pandas as pd

df = pd.DataFrame({'':new_words})
```

```python
df.to_csv("words.csv", index=False)
```
